```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Cardiovascular_Disease_Dataset.csv")
df.head(10)
```

```python
X = df.drop('target', axis=1)
y = df['target']
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
class LinearRegression:
    def __init__(self, learning_rate=0.001, n_iterations=1000):  # Adjusted learning rate
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # gradient descent here
        for _ in range(self.n_iterations):
            model = np.dot(X, self.weights) + self.bias
            dw = (1/n_samples) * np.dot(X.T, (model - y))
            db = (1/n_samples) * np.sum(model - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

```python
linregressor = LinearRegression()
linregressor.fit(X_train, y_train)
predictions = linregressor.predict(X_test)
```

```python
threshold = 0.5
binary_predictions = np.where(predictions > threshold, 1, 0)
```

```python
accuracy = accuracy_score(y_test, binary_predictions)
f1 = f1_score(y_test, binary_predictions)

# f1 score and accuracy
print("Accuracy:", accuracy)
print("F1 Score:", f1)
```
